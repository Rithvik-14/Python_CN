# The time module in Python-
provides a set of functions to work with time-related operations, such as timekeeping, formatting, and time conversions. 
Whether you are writing a script, a library, or an application, the time module is a powerful tool that can help you perform time-related tasks with ease and efficiency. 

```python
import time

def usingWhile():
    i=0
    while i< 200:
        print(i)
        i=i+1
        
def usingFor():
    for i in range(200):
        print(i)
        
init = time.time()  # it gives exact seconds since the time it has started.
usingFor()
t1 = time.time()-init
print(f'for loop takes: {t1} time')
time.sleep(3) # after 3 seconds it will run the next line.
init = time.time()
usingWhile()
print(f'while loop takes: {time.time() - init} time')

t = time.localtime()
formatted_time = time.strftime('%Y-%m-%d %H: %M: %S',t)  # strftime tells local time in the form of String.
print(formatted_time)
```

# Walrus Operater.
The Walrus Operator is represented by the := syntax and can be used in a variety of contexts including while loops and if statements.

```python
foods = list()
while True:
    food = input('what food do you like?')
    if food == 'quit':
        break
    else:
        foods.append(food)
```

```python
foods = list()
while( food := input('What food do you like? '))!='quit':
    foods.append(food)
```
